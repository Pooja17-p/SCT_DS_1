```python
import matplotlib.pyplot as plt

# Input data (in millions)
age_groups = ['0-10', '11-30', '31-50', '60+']
population = [400, 500, 350, 450]
```

```python
plt.figure(figsize=(8, 4))

# Blue color shades
colors = ['#3E2F5B','#5A4A82','#8A7FB5','#D8D3EA']

plt.barh(age_groups, population, color=colors)

plt.title('Age Group Population Distribution - 2023')
plt.xlabel('Population (in millions)')
plt.ylabel('Age Groups')

plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```
